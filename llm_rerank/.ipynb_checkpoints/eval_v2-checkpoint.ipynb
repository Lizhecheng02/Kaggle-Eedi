```python
import pandas as pd
pred_labels = list(pd.read_csv("submission_stage2.csv")["MisconceptionId"].apply(lambda x: [int(y) for y in x.split()]).values)
# pred_labels  = list(pd.read_csv("submission.csv")["MisconceptionId"].apply(lambda x: [int(y) for y in x.strip('[]').replace('\n', ' ').split()]))

# ttt = pd.read_parquet("submission_stage2_1.parquet").sort_values("QuestionId_Answer", ascending=True).reset_index(drop=True)
# true_labels = ttt["MisconceptionId"].values
true_labels = list(pd.read_parquet("submission_stage2_1_tmp.parquet")["MisconceptionId"].values)

def mean_average_precision(true_labels, predicted_labels_list, k=25):

    average_precisions = []
    tmp = []
    for true_label, predicted_labels in zip(true_labels, predicted_labels_list):
        if true_label in predicted_labels[:k]:
            idx = predicted_labels[:k].index(true_label)
            average_precisions.append(1/(idx + 1))
            tmp.append(1)
        else:
            average_precisions.append(0)
            tmp.append(0)
    print(sum(tmp) / len(tmp) if tmp else 0)

    return sum(average_precisions) / len(average_precisions) if average_precisions else 0
    
score = mean_average_precision(true_labels,pred_labels, 100)
print(score)
```

```python
rerank_data = pd.read_json('../datas/data/dev_data.json')
sub_df1= pd.read_csv("submission_stage2.csv")
sub_df1["recall_MisconceptionId"] = sub_df1["MisconceptionId"].apply(lambda x: [int(y) for y in x.strip('[]').replace('\n', ' ').split()])
del sub_df1["MisconceptionId"]


tmp = []
for i in range(len(rerank_data)):
    # text = f"ConstructName:{rerank_data.loc[i, 'ConstructName']}\nSubjectName:{rerank_data.loc[i, 'SubjectName']}\nQuestionText:{rerank_data.loc[i, 'QuestionText']}\nCorrectAnswerText:{rerank_data.loc[i, 'CorrectAnswerText']}\nInCorrectAnswerText:{rerank_data.loc[i, 'InCorrectAnswerText']}"
    text = f"ConstructName:{rerank_data.loc[i, 'ConstructName']}\nQuestionText:{rerank_data.loc[i, 'QuestionText']}\nCorrectAnswerText:{rerank_data.loc[i, 'CorrectAnswerText']}\nInCorrectAnswerText:{rerank_data.loc[i, 'InCorrectAnswerText']}"
    
    tmp.append(text)
    
rerank_data['text'] = tmp
rerank_data = rerank_data.merge(sub_df1, on='QuestionId_Answer', how='left')
rerank_data.head()
```

```python
m_map = pd.read_csv('../datas/misconception_mapping.csv')
m_datas = m_map['MisconceptionName'].tolist()


predict_list = []
TOP_NUM = 100
for _, row in rerank_data.iterrows():
    for index, m_id in enumerate(row['recall_MisconceptionId'][:TOP_NUM]):
        m_text = "Misconception:" + m_datas[int(m_id)]
        predict_list.append({
            "QuestionId_Answer": row['QuestionId_Answer'],
            'MisconceptionId': row['MisconceptionId'],
            'text': row['text'],
            'predict_text': [row['text'], m_text],
            'candi_MisconceptionId': m_id,
            'candi_MisconceptionName': m_text
        })
predict_list = pd.DataFrame(predict_list)
print(predict_list.shape)
predict_list.to_parquet("rerank_data_1.parquet", index=False)
predict_list.head()
```

```python
import numpy as np
predict_list = pd.read_parquet("rerank_data_res.parquet").sort_values("QuestionId_Answer", ascending=True).reset_index(drop=True)
predict_list.head()

predicts_test = []
for id, df in predict_list.groupby("QuestionId_Answer"):
    scores = df['score'].values
    score_indexs = np.argsort(-scores)
    candi_paper_ids = df['candi_MisconceptionId'].values
    origin_sorted = sub_df1_tmp[sub_df1_tmp.QuestionId_Answer == id].recall_MisconceptionId.tolist()[0][15:25]
    predicts_test.append([candi_paper_ids[index] for index in score_indexs[:25]])
    # predicts_test.append([candi_paper_ids[index] for index in score_indexs[:25] if candi_paper_ids[index] not in origin_sorted] +origin_sorted)
    

ttt = pd.read_parquet("submission_stage2_1_tmp.parquet").sort_values("QuestionId_Answer", ascending=True).reset_index(drop=True)
true_labels = ttt["MisconceptionId"].values
# true_labels = list(pd.read_parquet("submission_stage1.parquet")["MisconceptionId"].values)

def mean_average_precision(true_labels, predicted_labels_list, k=25):

    average_precisions = []
    tmp = []
    for true_label, predicted_labels in zip(true_labels, predicted_labels_list):
        if true_label in predicted_labels[:k]:
            idx = predicted_labels[:k].index(true_label)
            average_precisions.append(1/(idx + 1))
            tmp.append(1)
        else:
            average_precisions.append(0)
            tmp.append(0)
    print(sum(tmp) / len(tmp) if tmp else 0)

    return sum(average_precisions) / len(average_precisions) if average_precisions else 0
    
score = mean_average_precision(true_labels,predicts_test,25)
print(score)
```
